```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.cluster import KMeans





print("Libraries imported successfully!")
```

```python
X_train = pd.read_csv('./heart_disease_data/split_data/X_train_selected.csv')
X_test = pd.read_csv('./heart_disease_data/split_data/X_test_selected.csv')
y_train = pd.read_csv('./heart_disease_data/split_data/y_train.csv')['target_binary']
y_test = pd.read_csv('./heart_disease_data/split_data/y_test.csv')['target_binary']

print("*"*40, "All Data Loaded!!!", "*"*40)
```

```python
k_range = range(1, 11)
within_cluster_sum_squares = []  # Within-cluster sum of squares

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_train)
    within_cluster_sum_squares.append(kmeans.inertia_)  

print(within_cluster_sum_squares);
```

```python
plt.figure(figsize=(8, 5))
plt.plot(k_range, within_cluster_sum_squares, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares')
plt.grid(True, alpha=0.3)
plt.show()
```
