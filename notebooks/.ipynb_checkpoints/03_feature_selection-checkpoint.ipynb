```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import (
    chi2, RFE, SelectKBest
)
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression


print("Libraries imported successfully!")
```

```python
X_train = pd.read_csv('./heart_disease_data/split_data/X_train_scaled.csv')
X_test = pd.read_csv('./heart_disease_data/split_data/X_test_scaled.csv')
y_train = pd.read_csv('./heart_disease_data/split_data/y_train.csv')['target_binary']
y_test = pd.read_csv('./heart_disease_data/split_data/y_test.csv')['target_binary']

print("*"*40, "All Data Loaded!!!", "*"*40)
```

### Random Forest Feature Importance -> Looks at how often a feature is used in decision trees

```python
rf= RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
print("*"*40, "Training Done!", "*"*40, "\n")

importance= rf.feature_importances_
feature_names= X_train.columns

#Create importance dataframe 
feature_importance_df= pd.DataFrame(
    {
        'Feature': feature_names,
        'Importance' : importance
    }
).sort_values('Importance', ascending= False)


print("Top 10 Most Important Features:\n")
print(feature_importance_df.head(10))
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance_df.head(10), x='Importance', y='Feature', palette='viridis')
plt.title('Top 10 Feature Importance (Random Forest)')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.grid(True, axis='x', alpha=0.3)
plt.show()
```

### RFE (Recursive Feature Elimination) -> Removes features one by one, based on model performance

```python
model = LogisticRegression(max_iter=1000, random_state=42)

n_selection= 10 

rfe= RFE(model, n_features_to_select=n_selection)

X_train_rfe= rfe.fit_transform(X_train, y_train)

mask= rfe.support_
selected_features= X_train.columns[mask]


print("Top selected features by RFE:\n")
for i, feature in enumerate(selected_features, 1):
    print(f"  {i}. {feature}")
```

```python
rfe_ranking = rfe.ranking_

rfe_results = pd.DataFrame({
    'Feature': feature_names,
    'Rank': rfe_ranking
}).sort_values('Rank')

print("RFE Feature Rankings (Lower = More Important):\n")
print(rfe_results)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=rfe_results, y='Feature', x='Rank', palette='coolwarm', dodge=False)

# Customize
plt.xlabel('RFE Rank ')
plt.ylabel('Feature')
plt.title('Features Selected by RFE (Recursive Feature Elimination)')
plt.axvline(1.5, color='red', linestyle='--', alpha=0.5)  
plt.grid(True, axis='x', alpha=0.3)
plt.xlim(0, 2)



plt.show()
```

### Chi-Square Test -> Which categorical features are significantly linked to heart disease?


```python
X_train.info()
```

```python
categorical_features = ['sex', 'cp', 'fbs', 'exang', 'slope', 'ca', 'thal_unknown', 'thal_fixed_defect', 'thal_reversible_defect']

X_train[categorical_features]= X_train[categorical_features].astype("category")
X_train.info()
```

```python
X_train_cat = X_train[categorical_features]
X_train_cat
```

```python
chi2_selector= SelectKBest(score_func=chi2, k='all') 

X_train_chi2= chi2_selector.fit_transform(X_train_cat, y_train)  


chi2_scores= chi2_selector.scores_
p_values= chi2_selector.pvalues_ 


# Create a DataFrame
chi2_results = pd.DataFrame({
    'Feature': categorical_features,
    'Chi2_Score': chi2_scores,
    'P_Value': p_values
}).sort_values('Chi2_Score', ascending=False)

print("\nChi-Square Results:\n")
print(chi2_results)


print("\n")


for index, row in chi2_results.iterrows():
    if row['P_Value'] < 0.05:
        print(f"This {row['Feature']} is significant predictor.")
    else:
        print(f"This {row['Feature']} ----------------> is not significant predictor.")
    
   
```

```python
chi2_results_sorted = chi2_results.sort_values('Chi2_Score', ascending=True)

# Plot
plt.figure(figsize=(10, 6))
colors = ['red' if p < 0.05 else 'gray' for p in chi2_results_sorted['P_Value']]
sns.barplot(data=chi2_results_sorted, x='Chi2_Score', y='Feature', palette=colors)

plt.title('Chi-Square Test Results (Categorical Features)')
plt.xlabel('Chi-Square Score')
plt.ylabel('Feature')

# Add legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='red', label='Significant (p < 0.05)'),
                   Patch(facecolor='gray', label='Not Significant')]
plt.legend(handles=legend_elements, loc='lower right')

plt.grid(True, axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```
